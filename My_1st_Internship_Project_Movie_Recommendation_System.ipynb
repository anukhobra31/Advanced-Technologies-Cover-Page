```python
import random
movies = {
    "Action": ["Avengers",
    "Avengers Endgame",
    "The Dark Knight",
    "Batman Begins",
    "Man of Steel",
    "Justice League",
    "Gladiator",
    "Mad Max Fury Road",
    "John Wick",
    "John Wick Chapter 2",
    "John Wick Chapter 3",
    "Mission Impossible",
    "Mission Impossible Fallout",
    "Fast and Furious",
    "Furious 7",
    "Die Hard",
    "The Matrix",
    "300",
    "Terminator",
    "Terminator 2",
    "Predator",
    "Rambo",
    "Commando",
    "Taken",
    "The Expendables",
    "Sholay",
    "Dhoom",
    "Dhoom 2",
    "War",
    "Pathaan",
    "Tiger Zinda Hai",
    "Ek Tha Tiger",
    "Baahubali The Beginning",
    "Baahubali The Conclusion",
    "KGF Chapter 1",
    "KGF Chapter 2",
    "Ghajini",
    "Singham",
    "Dabangg",
    "Raees",
    "Don",
    "Don 2",
    "Krrish",
    "Krrish 3",
    "Gadar Ek Prem Katha"
    ],


    "Comedy": ["The Mask",
    "Home Alone",
    "Home Alone 2",
    "Superbad",
    "The Hangover",
    "The Hangover Part 2",
    "Dumb and Dumber",
    "Ace Ventura",
    "Mr Bean",
    "Mr Bean's Holiday",
    "Yes Man",
    "Bruce Almighty",
    "Charlie and the Chocolate Factory",
    "Jumanji",
    "Jumanji Welcome to the Jungle",
    "Central Intelligence",
    "We're the Millers",
    "21 Jump Street",
    "22 Jump Street",
    "Rush Hour",
    "Rush Hour 2",
    "Rush Hour 3",
    "Johnny English",
    "Johnny English Reborn",
    "Shaun of the Dead",
    "The Dictator",
    "Zombieland",
    "Ted",
    "Ted 2",
    "Pink Panther",
    "3 Idiots",
    "PK",
    "Hera Pheri",
    "Phir Hera Pheri",
    "Golmaal",
    "Golmaal Returns",
    "Golmaal 3",
    "Welcome",
    "Welcome Back",
    "Dhamaal",
    "Double Dhamaal",
    "Housefull",
    "Housefull 2",
    "Housefull 3",
    "Housefull 4",
    "Chup Chup Ke",
    "Munna Bhai MBBS",
    "Lage Raho Munna Bhai",
    "Bajrangi Bhaijaan",
    "Andaz Apna Apna",
    "Coolie No 1",
    "Partner",
    "Bhool Bhulaiyaa"
    ],


    "Drama": [
        "The Shawshank Redemption",
    "Forrest Gump",
    "Titanic",
    "The Green Mile",
    "A Beautiful Mind",
    "Fight Club",
    "The Pursuit of Happyness",
    "The Godfather",
    "The Godfather Part II",
    "Schindler's List",
    "12 Years a Slave",
    "Parasite",
    "Joker",
    "Black Swan",
    "The Social Network",
    "The Pianist",
    "Life of Pi",
    "Slumdog Millionaire",
    "Revolutionary Road",
    "The Curious Case of Benjamin Button",
    "Taare Zameen Par",
    "Swades",
    "Rang De Basanti",
    "Dil Chahta Hai",
    "Rockstar",
    "Kabir Singh",
    "Veer-Zaara",
    "Kal Ho Naa Ho",
    "My Name Is Khan",
    "Barfi",
    "Queen",
    "Highway",
    "Piku",
    "Dear Zindagi",
    "Udaan",
    "Masaan",
    "October",
    "Neerja",
    "Pink",
    "Bajrangi Bhaijaan",
    "Andhadhun",
    "Drishyam",
    "English Vinglish",
    "Haider",
    "A Wednesday"
    ],



    "Sci-Fi": [
       "Interstellar",
    "Inception",
    "Avatar",
    "Avatar The Way of Water",
    "The Matrix",
    "The Matrix Reloaded",
    "The Matrix Revolutions",
    "The Terminator",
    "Terminator 2 Judgment Day",
    "Blade Runner",
    "Blade Runner 2049",
    "Dune",
    "Dune Part Two",
    "Gravity",
    "Arrival",
    "Star Wars A New Hope",
    "Star Wars The Empire Strikes Back",
    "Star Wars Return of the Jedi",
    "Star Wars The Force Awakens",
    "Rogue One A Star Wars Story",
    "Alien",
    "Aliens",
    "Predator",
    "Jurassic Park",
    "Jurassic World",
    "E.T. the Extra-Terrestrial",
    "Back to the Future",
    "Back to the Future Part II",
    "Back to the Future Part III",
    "Looper",
    "Oblivion",
    "Edge of Tomorrow",
    "Minority Report",
    "I Am Legend",
    "War of the Worlds",
    "Ready Player One",
    "Ex Machina",
    "District 9",
    "Tron Legacy",
    "Krrish",
    "Krrish 3",
    "Ra One",
    "Robot",
    "Robot 2.0",
    "PK"
    ],



    "Horror": [
        "The Exorcist",
    "The Conjuring",
    "The Conjuring 2",
    "Annabelle",
    "Annabelle Creation",
    "The Nun",
    "Insidious",
    "Insidious Chapter 2",
    "Insidious Chapter 3",
    "Sinister",
    "Hereditary",
    "The Ring",
    "The Grudge",
    "It",
    "It Chapter Two",
    "A Quiet Place",
    "A Quiet Place Part II",
    "The Babadook",
    "The Omen",
    "Poltergeist",
    "The Amityville Horror",
    "Saw",
    "Saw II",
    "The Texas Chainsaw Massacre",
    "Halloween",
    "Paranormal Activity",
    "The Blair Witch Project",
    "Final Destination",
    "Final Destination 2",
    "Final Destination 3",
    "The Cabin in the Woods",
    "Evil Dead",
    "Doctor Sleep",
    "Lights Out",
    "Drag Me to Hell",
    "The Others",
    "Bhool Bhulaiyaa",
    "Bhool Bhulaiyaa 2",
    "Raaz",
    "Raaz 2",
    "1920",
    "1920 London",
    "Bhoot",
    "Bhoot Part One The Haunted Ship",
    "Stree",
    "Pari",
    "Mahal",
    "Tumbbad",
    "Darna Mana Hai"
    ],


"Romance": [
    "Titanic",
    "The Notebook",
    "La La Land",
    "Pride and Prejudice",
    "Before Sunrise",
    "Before Sunset",
    "Before Midnight",
    "A Walk to Remember",
    "Me Before You",
    "The Fault in Our Stars",
    "Notting Hill",
    "Pretty Woman",
    "Love Actually",
    "The Vow",
    "Five Feet Apart",
    "Call Me by Your Name",
    "Romeo and Juliet",
    "Blue Valentine",
    "The Time Traveler's Wife",
    "Eternal Sunshine of the Spotless Mind",
    "Crazy Rich Asians",
    "Silver Linings Playbook",
    "Her",
    "500 Days of Summer",
    "The Proposal",
    "Dilwale Dulhania Le Jayenge",
    "Veer-Zaara",
    "Kal Ho Naa Ho",
    "Kuch Kuch Hota Hai",
    "Jab We Met",
    "Aashiqui",
    "Aashiqui 2",
    "Rockstar",
    "Sanam Teri Kasam",
    "Barfi",
    "Tamasha",
    "Yeh Jawaani Hai Deewani",
    "Love Aaj Kal",
    "Hum Dil De Chuke Sanam",
    "Rehna Hai Tere Dil Mein",
    "Mohabbatein",
    "Fanaa",
    "Raanjhanaa",
    "Lootera",
    "Sita Ramam",
    "Gehraiyaan"
],



"Thriller":[
    "Se7en",
    "Gone Girl",
    "Shutter Island",
    "Prisoners",
    "The Silence of the Lambs",
    "Zodiac",
    "The Girl with the Dragon Tattoo",
    "Nightcrawler",
    "The Prestige",
    "Black Swan",
    "Memento",
    "The Sixth Sense",
    "Oldboy",
    "Fight Club",
    "No Country for Old Men",
    "A Quiet Place",
    "Get Out",
    "Split",
    "Glass",
    "The Usual Suspects",
    "The Departed",
    "Heat",
    "The Fugitive",
    "The Talented Mr. Ripley",
    "Argo",
    "Andhadhun",
    "Drishyam",
    "Drishyam 2",
    "Kahaani",
    "Badla",
    "Talvar",
    "Raman Raghav 2.0",
    "A Wednesday",
    "Special 26",
    "Ugly",
    "NH10",
    "Ittefaq",
    "Raazi",
    "Johnny Gaddaar",
    "Table No 21",
    "Darr",
    "Baazigar",
    "Ek Villain",
    "Ghajini",
    "Tees Maar Khan"
],


"Fantasy": [
    "Harry Potter and the Sorcerer's Stone",
    "Harry Potter and the Chamber of Secrets",
    "Harry Potter and the Prisoner of Azkaban",
    "Harry Potter and the Goblet of Fire",
    "Harry Potter and the Order of the Phoenix",
    "Harry Potter and the Half-Blood Prince",
    "Harry Potter and the Deathly Hallows Part 1",
    "Harry Potter and the Deathly Hallows Part 2",
    "The Lord of the Rings The Fellowship of the Ring",
    "The Lord of the Rings The Two Towers",
    "The Lord of the Rings The Return of the King",
    "The Hobbit An Unexpected Journey",
    "The Hobbit The Desolation of Smaug",
    "The Hobbit The Battle of the Five Armies",

    "Pirates of the Caribbean The Curse of the Black Pearl",
    "Pirates of the Caribbean Dead Man's Chest",
    "Pirates of the Caribbean At World's End",
    "Pirates of the Caribbean On Stranger Tides",
    "Pirates of the Caribbean Dead Men Tell No Tales",
    "Fantastic Beasts and Where to Find Them",
    "Fantastic Beasts The Crimes of Grindelwald",
    "Fantastic Beasts The Secrets of Dumbledore",
    "Alice in Wonderland",
    "Alice Through the Looking Glass",
    "The Chronicles of Narnia The Lion the Witch and the Wardrobe",
    "The Chronicles of Narnia Prince Caspian",
    "The Chronicles of Narnia The Voyage of the Dawn Treader",
    "Pan's Labyrinth",
    "Doctor Strange",
    "Doctor Strange in the Multiverse of Madness",
    "Thor",
    "Thor The Dark World",
    "Thor Ragnarok",
    "Avatar",
    "Avatar The Way of Water",
    "Maleficent",
    "Maleficent Mistress of Evil",
    "Baahubali The Beginning",
    "Baahubali The Conclusion",
    "Brahmastra",
    "Tumbbad",
    "Krrish",
    "Koi Mil Gaya"
],
}

def recommend_movies():
    genre = input("Enter genre: ").title()

    if genre in movies:
        count = int(input("How many movies do you want? "))
        available_movies = movies[genre]

        if count > len(available_movies):
            print("Not enough movies available.")
        else:
            print("Movie Recommendations:")
            for movie in random.sample(available_movies, count):
                print(movie)
    else:
        print("Genre not found.")

# add a new movie
def add_movie():
    genre = input("Enter genre: ").title()
    movie_name = input("Enter movie name: ").title()

    if genre in movies:
        movies[genre].append(movie_name)
    else:
        movies[genre] = [movie_name]

    print("Movie added successfully.")

# search a movie
def search_movie():
    search = input("Enter movie name to search: ").title()

    for genre in movies:
        if search in movies[genre]:
            print(f"Movie found in {genre} genre.")
            return

    print("Movie not found.")

while True:
    print("\n--- Movie Recommendation System ---")
    print("1. Get Movie Recommendations")
    print("2. Add a New Movie")
    print("3. Search for a Movie")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        recommend_movies()
    elif choice == "2":
        add_movie()
    elif choice == "3":
        search_movie()
    elif choice == "4":
        print("Thank you! Exiting program.")
        break
    else:
        print("Invalid choice. Try again.")
```
